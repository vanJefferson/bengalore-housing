```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
data = pd.read_csv('house prices\Bengaluru_House_Data.csv')
data.head()
```

```python
data.shape
```

```python
data.groupby('area_type')['area_type'].agg('count')
```

```python
new_data = data.drop(['area_type', 'availability', 'society', 'balcony'], axis='columns')
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
new_data_no_missing_values = new_data.dropna()
new_data_no_missing_values.head()
```

```python
new_data_no_missing_values['size'].unique()
```

```python
new_data_no_missing_values['bhk'] = new_data_no_missing_values['size'].apply(lambda x: int(x.split(' ')[0]))
new_data_no_missing_values.head()
```

```python
new_data_no_missing_values['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
new_data_no_missing_values[~new_data_no_missing_values['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
new_data_no_missing_values_copy = new_data_no_missing_values.copy()
new_data_no_missing_values_copy['total_sqft'] = new_data_no_missing_values_copy['total_sqft'].apply(convert_sqft_to_num)
new_data_no_missing_values_copy.head()
```

```python
new_data_no_missing_values_copy['price_per_sqft'] = new_data_no_missing_values_copy['price']*100000/new_data_no_missing_values_copy['total_sqft']
```

```python
new_data_no_missing_values_copy.head()
```

```python
new_data_no_missing_values_copy['location'].apply(lambda x: x.strip())
```

```python
location_stats = new_data_no_missing_values_copy.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats <= 10])
```

```python
location_less_than_10 = location_stats[location_stats <= 10]
location_less_than_10
```

```python
len(new_data_no_missing_values_copy['location'].unique())
```

```python
new_data_no_missing_values_copy['location'] = new_data_no_missing_values_copy['location'].apply(lambda x: 'other' if x in location_less_than_10 else x)
len(new_data_no_missing_values_copy['location'].unique())
```

```python
new_data_no_missing_values_copy.head(10)
```

```python
new_data_no_missing_values_copy[new_data_no_missing_values_copy['total_sqft']/new_data_no_missing_values_copy['bhk']<300]
```

```python
new_data_no_outliers = new_data_no_missing_values_copy[~(new_data_no_missing_values_copy['total_sqft']/new_data_no_missing_values_copy['bhk']<300)]
new_data_no_outliers.shape
```

```python
new_data_no_outliers['price_per_sqft'].describe()
```

```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        filtered_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, filtered_df], ignore_index=True)
    return df_out
    
new_data_no_outliers = remove_outliers(new_data_no_outliers)
new_data_no_outliers.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

new_data_no_bhk_outliers = remove_bhk_outliers(new_data_no_outliers)
new_data_no_bhk_outliers.shape
```

```python
new_data_no_bhk_outliers = new_data_no_bhk_outliers[new_data_no_bhk_outliers.bath<new_data_no_bhk_outliers.bhk+2]
new_data_no_bhk_outliers.shape
```

```python
cleaned_df = new_data_no_bhk_outliers.drop(['size', 'price_per_sqft'], axis='columns')
cleaned_df.head()
```

```python
dummies = pd.get_dummies(cleaned_df['location'])
dummies
```

```python
cleaned_df_with_dummies = pd.concat([cleaned_df, dummies.drop('other', axis='columns')], axis='columns')
cleaned_df_with_dummies.head()
```

```python
cleaned_df_with_dummies = cleaned_df_with_dummies.drop('location', axis='columns')
cleaned_df_with_dummies.head()
```

```python
cleaned_df_with_dummies.shape
```

```python
X = cleaned_df_with_dummies.drop('price', axis='columns')
X.head(3)
```

```python
y= cleaned_df_with_dummies.price
y.head(3)
```

```python
X.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
X_train.shape
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    
    print(x.shape)
    print(model.predict([x]))
    return model.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```
